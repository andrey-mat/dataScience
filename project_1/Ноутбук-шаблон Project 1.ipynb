```python
def trans_education(education:str) -> str:
    """Get first words from education field to shorted the education
        data to put it into short number of categories

    Args:
        education (str): string with raw education data

    Returns:
        str: short education description
    """
    result = education.split()
    try:
        result = ' '.join(result[:result.index('образование')]).lower()
    except ValueError:
        return education
    
    return result

# transforming the education data
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(trans_education)
print('Количество уникальных видов образования:', hh_data['Образование'].nunique())
print('Количество людей со школьным образованием: ', \
    hh_data[hh_data['Образование'] == 'среднее']['Образование'].count())

# drop the unnecessary fielf
hh_data.drop('Образование и ВУЗ', axis=1, inplace=True)

    
```

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(exp_data: str) -> int:
    """ Transform experience data to get work experience in months

    Args:
        exp_data (str): data string with experience 

    Returns:
        int: experience as number of months
    """
    #check if it NaN
    if exp_data != exp_data:
        return np.nan
    
    months, years = 0, 0
    exp_data = exp_data.split()[:6] # use only first 6 words

    for i in range(len(exp_data)):
        if 'месяц' in exp_data[i]:
            months = int(exp_data[i-1])
        if ('лет' in exp_data[i]) or ('год') in exp_data[i]:
            years = int(exp_data[i-1])
    
    result = years*12 + months
    if result:
        return result
    else:
        return np.nan

# transform the data in experience data field
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
print('Медианный опыт работы в месяцах: ', hh_data['Опыт работы (месяц)'].median())

# drop the experience field - we do not need it anymore
hh_data.drop('Опыт работы', axis=1, inplace=True)
```

```python
def get_city(input_data: str) -> str:
    """ Get city category

    Args:
        input_data (str): String with city / readiness to move/
                           business trip readiness data
    Returns:
        str: Moscow, SPb, 'million_city' or 'other' city
    """
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                      'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    city = input_data.split()[0].strip()
    if city in ['Москва', 'Санкт-Петербург']:
        return city
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
    

def get_readiness_to_move(input_data: str) -> bool:
    """ Get readiness to move status

    Args:
        input_data (str): String with city / readiness to move/
                           business trip readiness data
    Returns:
        bool: True if ready to move to another city, else False
    """
    input_data = input_data.split(',')
    for s in input_data:
        if 'перее' in s:
            if 'не' in s:
                return False
            else:
                return True


def get_readiness_for_business_trips(input_data: str) -> bool:
    """ Get business trips readiness

    Args:
        input_data (str): String with city / readiness to move/
                           business trip readiness data

    Returns:
        bool: True if ready for business trips, else False
    """
    # check if is NaN
    if str != str:
        return False
    input_data = input_data.split(',')
    for s in input_data:
        if 'команд' in s:
            if 'не' in s:
                return False
            else:
                return True
    # in other cases - not ready
    return False

# data transformation
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_city)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_readiness_to_move)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_readiness_for_business_trips)

display(hh_data['Город'].value_counts(normalize=True).round(2)*100)

# get ready for business trip and ready to move people
fully_ready = hh_data['Готовность к командировкам'] * hh_data['Готовность к переезду']
print('---')
print('Готовы к переездам и к командировкам:')
display(fully_ready.value_counts(normalize=True).round(2)*100)

# drop the used data
hh_data.drop('Город, переезд, командировки', axis=1, inplace=True)
del fully_ready
```

5) Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:

* полный день, удаленная работа

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_occupation(occupation: str) -> dict:
    """Get occupation information. Splits the string into list of sentences and returns the dict 
        with each sentence count.

    Args:
        occupation (str): occupation string

    Returns:
        dict: dict with all types of available occipation status
    """
    from collections import Counter
    occupation = [s.strip() for s in occupation.split(',')]
    return dict(Counter(occupation))

#get the dicts with occupation data
hh_data['temp'] = hh_data['Занятость'].apply(get_occupation)
occupations = ['полная занятость', 'частичная занятость',
                'проектная работа', 'стажировка', 'волонтерство']

for occupation in occupations:
    hh_data[occupation] = hh_data['temp'].apply(
        lambda d: bool(d.get(occupation)) if d.get(occupation) else False)
    
#get the dicts with type of work
types_of_work = ['полный день', 'сменный график',
                 'гибкий график', 'удаленная работа', 'вахтовый метод']
hh_data['temp'] = hh_data['График'].apply(get_occupation)
for work in types_of_work:
    hh_data[work] = hh_data['temp'].apply(
        lambda d: bool(d.get(work)) if d.get(work) else False)

#gathering requested data
proj_vol = hh_data['проектная работа'] * hh_data['волонтерство']
print('Количество кандидатов, согласных на проектную работу и волонтерство: ', proj_vol.sum(),'candidates')
shift_flex = hh_data['вахтовый метод'] * hh_data['гибкий график']
print('Вахтовый метод и гибкий график предпочитают: ', shift_flex.sum(),'candidates')
#clean the data
hh_data.drop(['temp', 'График', 'Занятость'], axis=1, inplace=True)
del proj_vol, shift_flex
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#read exchange rates data
exchange_data = pd.read_csv('data/ExchangeRates.csv')
exchange_data.head()
#bring all date columns to datetime type
hh_data['Дата обновления резюме'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date
exchange_data['date'] = pd.to_datetime(exchange_data['date']).dt.date

#split the salary column
def salary_split(salary:str) -> tuple:
    """Returns the tuple with salary and currency

    Args:
        salary (str): salary with currency

    Returns:
        tuple: int salary and str currency in one tuple
        
    """
    salary = salary.split()
    salary_num = int(salary[0])
    cur_dict = {'грн': 'UAH',
                'USD': 'USD',
                'EUR': 'EUR',
                'бел.руб': 'BYN',
                'KGS': 'KGS',
                'сум': 'UZS',
                'AZN': 'AZN',
                'KZT': 'KZT',
                'руб': 'RUB'
    }
    for key, value in cur_dict.items():
        if key in salary[1]:
            cur_num = value
            break
    
    return salary_num, cur_num
    
    
hh_data['ЗП в валюте'] = hh_data['ЗП'].apply(
    salary_split).apply(
        lambda t: t[0])
hh_data['Валюта ЗП'] = hh_data['ЗП'].apply(
    salary_split).apply(
        lambda t: t[1])

# merge cv data and exchange data
hh_data = hh_data.merge(
    exchange_data[['date', 'currency', 'proportion', 'close']], 
    how='left',
    left_on=['Валюта ЗП', 'Дата обновления резюме'],
    right_on=['currency', 'date'],
    )
# specific rate 1:1 for RUB
hh_data.loc[(hh_data['Валюта ЗП'] == 'RUB'), ['close', 'proportion']] = 1

# calculate salary in rubles
hh_data['ЗП (руб)'] = hh_data['ЗП в валюте'] * hh_data['close'] / hh_data['proportion']
print('Медианная зарплата:', round(hh_data['ЗП (руб)'].median() / 1000), 'тысяч рублей')

#cleaning data
hh_data.drop(['Валюта ЗП', 'ЗП в валюте', 
              'close', 'proportion', 'date',
              'currency', 'Дата обновления резюме', 'ЗП'],
             axis=1, 
             inplace=True)
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
hh_data['Age_log'] = np.log(hh_data['Возраст'])

hist = sns.histplot(
    data=hh_data,
    x='Age_log',
    bins=30,
    ax=axes,
)
hist.set_title('Возраст - логарифмическое распределение')

mu = hh_data['Age_log'].mean()
sigma = hh_data['Age_log'].std()
upper_bound = mu + 4*sigma
lower_bound = mu - 3*sigma

hist.axvline(mu, color='k', lw=2)
hist.axvline(upper_bound, color='k', lw=2, ls='--')
hist.axvline(lower_bound, color='k', lw=2, ls='--')

age_outliers = hh_data[(hh_data['Age_log'] < lower_bound) | (hh_data['Age_log'] > upper_bound)]
print('Количество отклонений (3 сигмы влево и 4 вправо)', age_outliers.shape[0])
print('Отклонения:')
print(age_outliers)
```
